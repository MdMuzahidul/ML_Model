```python
from pymongo import MongoClient
import pandas as pd
```

### Creat MongoDB connection URI and Access database and collection


```python
uri = "mongodb://localhost:27017"
client = MongoClient(uri)

db = client["Books_Recommendation"]
collection = db["Books"]
```

```python
data = collection.find()
books = pd.DataFrame(data)

# drop item if any null value in the row
books = books.dropna()

books.info()
```

### title series author rating description genres characters awards numRatings

```python
books=books[["isbn","title","series","author","rating","description","genres","characters","numRatings"]]
books.info()
```

```python
books.head()
```

```python
books['genres']
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
books['genres'] = books['genres'].apply(collapse)
books['genres']
```

```python
books['characters'] = books['characters'].apply(collapse)
books['characters']
```

```python
books['author'] = books['author'].apply(lambda x: x.replace(" ",""))
books['author']
```

```python
books.head(10)
```

```python
books['description'] = books['description'].apply(lambda x: x.replace("\n", " ") if isinstance(x, str) else "")
books['description']
```

```python
books["title"] = books["title"].apply(lambda x: x.replace(" ",""))
books["title"]
```

```python
books['tags'] = (
    books['title'].apply(lambda x: [x])+
    books['author'].apply(lambda x: [x]) +
    books['series'].apply(lambda x: [x] if isinstance(x, str) else []) +
    books['genres'] +
    books['characters'] +
    books['description'].apply(lambda x: x.split()) +
    books['rating'].apply(lambda x: [str(x)]) 
)
books['tags']
books.head(10)
```

```python
new_tags = books.drop(columns=['series', 'author', 'rating', 'description', 'genres', 'characters', 'numRatings'])
```

```python
new_tags['tags'] = new_tags['tags'].apply(lambda x: " ".join(x))
new_tags.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Remove rows with empty or NaN tags
new_tags = new_tags[new_tags['tags'].notnull() & (new_tags['tags'] != '')]

# Now vectorize
vector = cv.fit_transform(new_tags['tags']).toarray()

# Then compute similarity
similarity = cosine_similarity(vector)
```

```python
vector.shape, similarity.shape
```

```python
new_tags[new_tags['title'] == 'Dracula'].index[0]
```

```python
def recommend(book):
    index = new_tags[new_tags['title'] == book].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_tags.iloc[i[0]].title)
```

```python
recommend('Dracula')
```
